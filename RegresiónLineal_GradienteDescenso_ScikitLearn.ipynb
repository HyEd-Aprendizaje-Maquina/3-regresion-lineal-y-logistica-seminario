```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar el conjunto de datos de Advertising
url = "https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv"
data = pd.read_csv(url)
data.head(10)
```

```python
# Seleccionar las columnas relevantes (TV y Sales)
X = data[['TV']].values  # Gasto en publicidad en TV
y = data[['sales']].values  # Ventas generadas

# Normalizar los datos
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X_scaled, y_scaled)

# Predicciones
predicciones = modelo.predict(X_scaled)

# Graficar los datos y la recta de regresi칩n
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled, y_scaled, color='blue', label='Datos reales')
plt.plot(X_scaled, predicciones, color='red', linewidth=2, label='Recta de regresi칩n')
plt.title('Regresi칩n Lineal (Publicidad en TV vs Ventas)')
plt.xlabel('Publicidad en TV (normalizado)')
plt.ylabel('Ventas (normalizado)')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar los par치metros del modelo
print(f"Intercepto: {modelo.intercept_[0]}")
print(f"Pendiente: {modelo.coef_[0][0]}")
```
